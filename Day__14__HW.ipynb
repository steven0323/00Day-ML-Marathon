```python
import numpy as np
np.random.seed(1)
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

# 負相關
x = np.random.randint(0,50,1000)
y = np.random.randint(0,50,1000)
np.corrcoef(x,y)

plt.scatter(x,y)

#正相關
a = np.random.randint(0,50,1000)
b = a + np.random.normal(0,10,1000)
np.corrcoef(a,b)

plt.scatter(a,b)
```
