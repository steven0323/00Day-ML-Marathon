```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')

%matplotlib inline
import warnings
warnings.filterwarnings('ignore')


#homework-1 
matrix = np.random.uniform(-1,1,(10,10))
#print(matrix)
plt.figure(figsize = (10,10))
sns.heatmap(matrix,cmap = plt.cm.RdYlBu_r,vmin= -1,annot = True,vmax = 1)
plt.title('Homework-1')
plt.show()

```

```python
#Homework-2
nrow = 1000
ncol = 3
matrix = np.random.uniform(-1,1,(1000,3))
indice = np.random.choice([0,1,2],size=nrow)
plot_data = pd.DataFrame(matrix,indice).reset_index()

grid = sns.PairGrid(data=plot_data, size=3,diag_sharey=False,hue='index',vars = [x for x in list(plot_data.columns) if x!='index'])
grid.map_upper(plt.scatter,alpha=0.2)
grid.map_diag(sns.kdeplot)
grid.map_lower(sns.kdeplot,cmap=plt.cm.OrRd_r)
plt.suptitle('homework-2',size=32,y=1.05)
plt.show()
```
