```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

#read in file
f_app = os.path.expanduser('~/Desktop/application_train.csv')
app_train = pd.read_csv(f_app)

for col in app_train:
    if app_train[col].dtype == 'object':
        
        if len(list(app_train[col].unique())) <=2:
            app_train[col] = le.fit_transform(app_train[col])
app_train.head()

app_train['DAYS_EMPLOYED_ANOM'] = app_train['DAYS_EMPLOYED'] == 365243
app_train['DAYS_EMPLOYED'].replace({365243:np.nan},inplace = True)

app_train['DAYS_BIRTH'] = abs(app_train['DAYS_BIRTH'])
print(app_train['AMT_INCOME_TOTAL'].max())
            
```

```python
print(len(app_train['AMT_INCOME_TOTAL']))
app_train['equal_width_income'] = pd.cut(app_train['AMT_INCOME_TOTAL'],25)
app_train['equal_width_income'].value_counts()
plt.hist(app_train['AMT_INCOME_TOTAL']/10000,edgecolor='k')
plt.title('amt_income');plt.xlabel('set_division');plt.ylabel('income value')
plt.show()
```
